```python
from spextra import SpecLibrary, Spextrum, Passband
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
import astropy.units as u
```

```python
### HIGH TEMPERATURE BLACKBODY ###

starTemp = 5800
starMag = 5
filterName ='etc/V'
mirrorArea = 0.5

blackBodySpec = Spextrum.black_body_spectrum(temperature = starTemp, amplitude = starMag*u.ABmag, filter_curve=filterName)
starPhotons = blackBodySpec.photons_in_range(area=mirrorArea,filter_curve=filterName)
blackBodySpec.plot()

print(starPhotons)
print(blackBodySpec)
```
